```python
params = { 'year' : '2016', 'query_text': 'fire'
          
}

```

```python
fire = requests.get(endpoint, params = params)
fire.text
```

# this changes it to a dataframe

```python
fire = fire.json()
fire_2016 = pd.DataFrame.from_dict(fire)
```

```python
fire_2016.head(5)
```

```python
print(fire_2016)
```

to use n.largest you must have a float64 or int

```python
fire_2016 = fire_2016.astype({'query_count':'float'})
fire_2016.nlargest(1,'query_count')

```
